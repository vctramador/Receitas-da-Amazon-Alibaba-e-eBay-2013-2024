```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Alibaba.csv")
df
```

```python
# limpando valores do dataframe
df = df.dropna()
df
```

```python
# analisando os maiores valores
df_ordem1 = df.sort_values(by="Billion CYD", ascending=False)
df_ordem1
```

```python
# analisando a tendÃªncia de crescimento ou queda com o passar dos anos
df_ordem = df.sort_values(by="Year", ascending=True)
df_ordem
```

```python
#analisando o crescimento em porcentagens
df['Growth Rate (%)'] = df['Billion CYD'].pct_change() * 100
df['Growth Rate (%)'] = df['Growth Rate (%)'].fillna(0)
df
```

```python
# convertendo valores em variaveis para a criaÃ§Ã£o do grÃ¡fico
df_y = df['Year']
df_x = df['Billion CYD']
```

```python
plt.figure(figsize=(15, 9))
plt.bar(df_y, df_x)
plt.xlabel('Year')
plt.xticks(rotation=90) # ajustando a rotaÃ§Ã£o para evitar sobreposiÃ§Ã£o
plt.tight_layout()
plt.ylabel('BilhÃµes em CYD')
plt.title('GrÃ¡fico de Barras')
plt.show()
```

```python
# salvando o arquivo em csv
df.to_csv('Alibaba_alterado.csv', index=False)
```
