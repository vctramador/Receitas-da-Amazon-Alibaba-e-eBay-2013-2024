```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("eBay.csv")
df
```

```python
# limpando valores do dataframe
df = df.dropna()
df
```

```python
# analisando os maiores valores
df_ordem1 = df.sort_values(by="Billion USD", ascending=False)
df_ordem1
```

```python
# analisando a tendência de crescimento ou queda com o passar dos anos
df_ordem = df.sort_values(by="Year", ascending=True)
df_ordem
```

```python
#analisando o crescimento em porcentagens
df['Growth Rate (%)'] = df['Billion USD'].pct_change() * 100
df['Growth Rate (%)'] = df['Growth Rate (%)'].fillna(0)
df
```

```python
# convertendo valores em variaveis para a criação do gráfico
df_y = df['Year']
df_x = df['Billion USD']
```

```python
plt.figure(figsize=(15, 9))
plt.bar(df_y, df_x)
plt.xlabel('Year')
plt.xticks(rotation=90) # ajustando a rotação para evitar sobreposição
plt.tight_layout()
plt.ylabel('Bilhões em USD')
plt.title('Gráfico de Barras')
plt.show()
```

```python
# salvando o arquivo modificado em csv
df.to_csv('eBay_modificado.csv', index=False)
```
